```python
import pandas as pd
import numpy as np
import json

```

```python
movieId_lookup = pd.read_csv('../dataset/processed/movieId_lookup.csv')
tmdb_movies_info =pd.read_csv('../dataset/processed/tmdb_movies_info.csv')
movieLens_movies_info =pd.read_csv('../dataset/MovieLens/movies.csv')
with open('../dataset/processed/userWatched_movies.json','r') as file:
    userWatched_movies = json.load(file)
```

```python
movie_ratings = movieId_lookup = pd.read_csv('../dataset/MovieLens/ratings.csv')
```

```python
userWatched_movies = pd.Series(userWatched_movies)
```

```python
movieId_movieLense = set()
for user_id,movieId_list in userWatched_movies.items():
    for movieId in movieId_list:
        movieId_movieLense.add(movieId)
```

```python
movie_matched = movieId_lookup[movieId_lookup['movieLens_movieId'].isin(movieId_movieLense)]
# movie_map_table =  dict(zip(movie_matched['movieLens_movieId'],movie_matched['tmdb_movieId']))
```

```python
movie_genre_keyword_info = movie_matched.merge(tmdb_movies_info[['movie_id','keywords_clean','genres_clean']],left_on='tmdb_movieId',right_on='movie_id',how='left')
```

```python
movie_genre_keyword_info = movie_genre_keyword_info[['title_clean','movieLens_movieId','keywords_clean','genres_clean']]
```

```python
watched_userId =userWatched_movies.keys().tolist()
```

```python
watched_userId = [int(userId) for userId in watched_userId]
```

```python
filter_ratings = movie_ratings[movie_ratings['userId'].isin(watched_userId)]
```
