```python
import pandas as pd
import numpy as np
import json

from collections import Counter

```

```python
movieId_lookup = pd.read_csv('../dataset/processed/movieId_lookup.csv')
tmdb_movies_info =pd.read_csv('../dataset/processed/tmdb_movies_info.csv')
movieLens_movies_info =pd.read_csv('../dataset/MovieLens/movies.csv')
movie_ratings = pd.read_csv('../dataset/MovieLens/ratings.csv')
```

```python
movie_ratings = movieId_lookup = pd.read_csv('../dataset/MovieLens/ratings.csv')
```

```python
# open userWatched movies json file.
with open('../dataset/processed/userWatched_movies.json','r') as file:
    userWatched_movies = json.load(file)
userWatched_movies = pd.Series(userWatched_movies)
```

```python
# save the rating of movies which are available in both databases.
exist_moives_only = movie_ratings.query(f'movieId in {movieId_lookup['movieLens_movieId'].unique().tolist()}')
```

```python
exist_moives_only.to_csv('../dataset/processed/exist_moives_rating.csv',index=False)
```

```python
def get_user_data(user_id):
        
    user_preference_info = {}
    
    user_ratings = movie_ratings.query(f'userId ==  {user_id}')
    user_ratings_moviesInfo = user_ratings[['movieId','rating']].merge(movieLens_movies_info,on='movieId',how='left')
    user_top_liked_movies = user_ratings_moviesInfo.loc[user_ratings_moviesInfo['rating'].sort_values(ascending=False).index][:15]
    
    user_liked_genres_list = user_top_liked_movies['genres'].str.split('|').tolist()
    
    user_liked_genres = [genre for movies_genres in user_liked_genres_list for genre in movies_genres]
    
    
    
    top_genres_liked = pd.Series(user_liked_genres).value_counts(ascending=False)[:4].index.tolist()
    user_preference_info['top_genres'] = top_genres_liked

    user_preference_info['average_rating'] = user_ratings['rating'].mean()

    # top favorite movies
    user_favorite_movies = user_top_liked_movies[:10]['title'].tolist()
    user_preference_info['favorite_movies'] = user_favorite_movies

    return user_preference_info
```

```python
user_movies_historical_data = []
```

```python

for user_idd in userWatched_movies.index.to_list():
    current_user = {'user_id':user_idd}
    information_user = get_user_data(user_idd)
    current_user.update(information_user)
    user_movies_historical_data.append(current_user)

    
    # print(f'information about {user_idd}:')
    # print(f'top_genres : {information_user['top_genres']}')
    # print(f'average_rating : {information_user['average_rating']}')
    # print(f'favorite movies : {information_user['favorite_movies']}')

    
    # print(f'*'*30)
```

```python
# store it for explaining user behaviour in frontend.
with open('../dataset/processed/user_movies_historical_data.json','w') as file:
    json.dump(user_movies_historical_data,file,indent=2)
```
