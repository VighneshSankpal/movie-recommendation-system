```python
import requests

# url = "https://image.tmdb.org/t/p/w780/6BVCgmhLeSTF8nPBuE9ncHHcCNv.jpg"
url ='https://image.tmdb.org/t/p/w780/bCYuM18r8G8pOlHf6yYxnFU9IJ4.jpg'
file_path = "movie_poster3.jpg"

# 1. Send a GET request with stream=True for efficient downloading
response = requests.get(url, stream=True)

# 2. Check if the request was successful
if response.status_code == 200:
    # 3. Open a local file in binary write mode ('wb')
    with open(file_path, "wb") as image_file:
        # 4. Write the image data in chunks to prevent high memory usage
        for chunk in response.iter_content(chunk_size=8192):
            image_file.write(chunk)
    print(f"Success! Image saved to {file_path}")
else:
    print(f"Failed to download image. Status code: {response.status_code}")
```

```python
def movie_card(movie_id):
    url =f'https://api.themoviedb.org/3/movie/{movie_id}?api_key={API_KEY}'
    response = requests.get(url,timeout=10)
    data = response.json()
    
    poster_url =  f'https://image.tmdb.org/t/p/w780{data["poster_path"]}'

    
    poster_response = requests.get(poster_url, stream=True)

    return poster_response.status_code
```
