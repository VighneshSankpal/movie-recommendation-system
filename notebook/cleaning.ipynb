```python
import pandas as pd
import numpy as np
import os

import json
```

```python
PATH_MOVIELENS = os.path.join('..','dataset','TMDB 5000')
```

```python
movies = pd.read_csv(os.path.join(PATH_MOVIELENS,'tmdb_5000_credits.csv'),usecols=['movie_id','cast','crew'])
movie_info = pd.read_csv(os.path.join(PATH_MOVIELENS,'tmdb_5000_movies.csv'))
movies.head()
```

```python
movie_info.head()
```

```python
movies_df = movies.merge(right=movie_info,left_on='movie_id',right_on='id',how='left')
```

```python
movies_df.drop(columns='id',inplace=True)
movies_df.head()
```

```python
movies_df.shape
```

```python
# select released movies only.
movies_df = movies_df[movies_df['status'] == 'Released']
```

```python
# content based filtering

# term frequency-inverse document frequency

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

# replace nan with empty string
movies_df['overview_clean'] = movies_df['overview'].fillna('')

# construct the TF-IDF matrix
tfidf_matrix = tfidf.fit_transform(movies_df['overview_clean'])

tfidf_matrix.shape
```

```python
def process_genres(x):
    x = json.loads(x)
    genres = {g['name'] for g in x}
    result = " "+ ', '.join(genres)
    return result

# genres_clean =  movies_df['genres'].apply(process_genres)
movies_df['genres_clean'] =  movies_df['genres'].apply(process_genres)
movies_df['genres_clean'].head(2)
```

```python
# cast
def process_cast(x):
    '''
    x : input record
    '''
    main_cast = list(range(0,4))
    x = json.loads(x)
    cast_members = list()
    for cast in x:
        if cast['order'] in main_cast:
            cast_members.append(cast['name'].strip())

    result = " " + ', '.join(cast_members)
    return result
movies_df['cast_clean'] = movies_df['cast'].apply(process_cast)
```

```python
# crews

def process_crew(x):

    x = json.loads(x)
    
    # imp_crew = ['Director','Writer', 'Cinematographer']
    # crew_members = []
    # director = ''
    result = ' '
    for crew in x:
        if crew['job'] == 'Director':
            # crew_members.append(f'{crew['job']} is {crew['name']}')
            result = f'Director: {crew['name']}'
            

    # result = " " + ', '.join(crew_members)
    return result
movies_df['crew_clean'] = movies_df['crew'].apply(process_crew)
```

```python
# keywords
def process_keywords(x):

    x  = json.loads(x)
    keywords = {k['name'] for k in x}

    result = " " + ', '.join(keywords)
    return result

movies_df['keywords_clean'] = movies_df['keywords'].apply(process_keywords)
```
